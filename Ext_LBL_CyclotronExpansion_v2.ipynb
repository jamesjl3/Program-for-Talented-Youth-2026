### The challenge

The 88-Inch Cyclotron at Lawrence Berkeley National Laboratory is named for the diameter of its magnet poles, the physical size of the machine that bends and accelerates charged particles in a spiral path. It is also known by a second, more precise name in accelerator physics: it is a **K=140 cyclotron**.

Every cyclotron has a K-value, a single number that sets the ceiling on how much energy per nucleon it can deliver to any beam:

$$\frac{E}{A} = K \left(\frac{Q}{A}\right)^2$$

where E is the beam's total kinetic energy, A is the ion's mass number, and Q is its charge state (how many electrons have been stripped off). Look closely at this formula: there are **two separate ways** to increase a beam's energy. You can increase K, which means a bigger magnet (radius) or stronger field. Or you can increase Q, the charge state, which means a better ion source, with no change to the magnet at all.

**Your group's real interest is the second lever: increasing the charge state of the incoming particles.** Since energy depends on Q squared, this can be an even bigger effect than a magnet upgrade, and it is achieved by upgrading the ECR (Electron Cyclotron Resonance) ion source rather than the cyclotron itself. This notebook walks through both levers so you can compare them directly, but Parts 3 and 4 are built specifically around your group's charge-state question.

$$K = (\text{constant}) \times B^2 r^2$$

In this notebook you will calculate the K-value scaling first (as a baseline comparison), apply the formula to a real published beam energy from the 88-Inch's own specifications, and then focus on what increasing the charge state specifically buys you, and what it would take to achieve it.

```python
import numpy as np
import matplotlib.pyplot as plt

print("Libraries loaded!")
```

## Part 1: Baseline Comparison — The Radius Lever

Before focusing on charge state, it's worth knowing what the *other* lever would give you, as a comparison point. The current 88-Inch Cyclotron has K = 140. If B stays the same and the radius grows from 88 to 100 inches, the new K-value scales as the square of the radius ratio, since K is proportional to B^2 r^2 and B is unchanged.

```python
K_88 = 140   # the 88-Inch Cyclotron's real, published K-value

r_ratio = (100 / 88)**2
K_100 = K_88 * r_ratio

print(f"Radius ratio squared, (100/88)^2: {r_ratio:.4f}")
print(f"Current K-value (88-inch):  K = {K_88}")
print(f"Upgraded K-value (100-inch): K = {K_100:.1f}")
print()
print(f"This is a real, meaningful accelerator rating jump: roughly K=140 to K={K_100:.0f},")
print("comparable in scale to the naming convention used at other real cyclotron facilities")
print("(for example, K500 and K1200 class machines elsewhere in the field).")
```

---

## Part 2: A Real Beam, A Real Number

The 88-Inch Cyclotron's own published specifications state that a bismuth-209 beam (Z = 83, A = 209) currently reaches a maximum energy of 4.5 MeV per nucleon. We use this real number to see exactly what the upgrade buys, rather than an invented example.

```python
# ── Real published fact ─────────────────────────────────────────────────────
Z_bi, A_bi = 83, 209
E_per_A_current = 4.5   # MeV/nucleon, real published maximum for Bi-209 at the 88-Inch

E_total_current = E_per_A_current * A_bi
print(f"Bismuth-209 beam, current (K={K_88}):")
print(f"  Energy per nucleon: {E_per_A_current} MeV")
print(f"  Total kinetic energy: {E_total_current:.1f} MeV")

# ── Solve for the implied charge state at the current K-value ─────────────
Q_over_A = np.sqrt(E_per_A_current / K_88)
Q_implied = Q_over_A * A_bi
Q_round = int(np.ceil(Q_implied))   # round up: a charge state must be a whole number,
                                      # and rounding up is the conservative choice

print(f"\n  Implied charge state: Q/A = {Q_over_A:.4f}, Q = {Q_implied:.1f}, so Q = {Q_round}+")
print("  A charge state in this range is realistic for the type of high-charge-state")
print("  ECR ion source the 88-Inch uses, which is a good consistency check on the formula.")
```

```python
# ── Same charge state, new K-value ──────────────────────────────────────────
# The ECR ion source does not change in this upgrade, only the cyclotron radius does,
# so we hold Q fixed and recompute E/A at the new K-value.

E_per_A_new = K_100 * (Q_round / A_bi)**2
E_total_new = E_per_A_new * A_bi

print(f"Bismuth-209 beam, upgraded (K={K_100:.0f}), same Q={Q_round}+:")
print(f"  New energy per nucleon: {E_per_A_new:.2f} MeV")
print(f"  New total kinetic energy: {E_total_new:.1f} MeV")
print()
print(f"Change: {E_total_current:.1f} MeV -> {E_total_new:.1f} MeV, "
      f"a {(E_total_new/E_total_current - 1)*100:.1f}% increase, without changing the ion source at all.")

fig, ax = plt.subplots(figsize=(7, 4.5))
ax.bar(['Current\n(88-inch)', 'Upgraded\n(100-inch)'], [E_total_current, E_total_new],
       color=['steelblue', 'firebrick'], alpha=0.85, width=0.5)
ax.set_ylabel('Total kinetic energy (MeV)', fontsize=12)
ax.set_title(f'Bismuth-209 beam energy, same charge state (Q={Q_round}+)', fontsize=12)
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

## Part 3: Increasing the Charge State Instead

Now the lever your group actually cares about. Instead of changing the radius (and therefore K), we hold K fixed at its current value of 140 and ask: what if the ion source could strip off more electrons, raising the charge state Q?

The bismuth beam currently runs at Q = 38+ (the value we solved for in Part 2). A real, cited motivation for pushing charge states higher: accelerator groups building on the 88-Inch's ECR source have proposed superconducting ECR designs driven by higher-frequency microwave sources specifically to reach higher charge states and energies, without touching the magnet at all.

```python
# ── Holding K fixed at 140, explore a range of higher charge states ─────────
Q_values = np.arange(Q_round, Q_round + 21)   # from current Q up to 20 higher

E_per_A_vs_Q = K_88 * (Q_values / A_bi)**2
E_total_vs_Q = E_per_A_vs_Q * A_bi

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(Q_values, E_total_vs_Q, marker='o', markersize=5, color='steelblue', linewidth=2,
        label='Energy vs charge state (K held fixed at 140)')
ax.axhline(y=E_total_current, color='gray', linestyle=':', linewidth=1.5,
           label=f'Current energy at Q={Q_round}+ ({E_total_current:.0f} MeV)')
ax.axhline(y=E_total_new, color='firebrick', linestyle='--', linewidth=1.5,
           label=f'Radius-upgrade energy at K={K_100:.0f} ({E_total_new:.0f} MeV)')
ax.set_xlabel('Charge state Q (+)', fontsize=12)
ax.set_ylabel('Total kinetic energy (MeV)', fontsize=12)
ax.set_title('Bismuth-209 beam energy vs charge state, same K-value', fontsize=12)
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ── What charge state would match the radius upgrade's energy gain? ────────
Q_match = A_bi * np.sqrt(E_total_new / A_bi / K_88)
print(f"To match the radius upgrade's energy ({E_total_new:.0f} MeV) by raising charge state alone,")
print(f"the ion source would need to reach Q = {Q_match:.1f}+, up from the current Q = {Q_round}+.")
print(f"That is an increase of only {Q_match - Q_round:.1f} charge units to match a {(r_ratio-1)*100:.1f}% radius increase.")
```

---

## Part 4: What Would It Take to Reach a Higher Charge State?

Charge state is set entirely by the ion source, not the cyclotron magnet. The 88-Inch uses ECR (Electron Cyclotron Resonance) ion sources, which strip electrons off atoms using microwave heating inside a magnetic bottle. Reaching a higher charge state generally requires:

- **Higher microwave frequency**, which allows a stronger confining magnetic field inside the ion source itself (a separate, much smaller magnet than the cyclotron's own bending magnet)
- **Longer confinement time**, giving electrons more chances to be stripped away before the ions are extracted
- **More microwave power**, to keep ionizing electrons out of a hotter, more energetic plasma

This is a real upgrade path pursued at ECR facilities generally: designs for next-generation superconducting ECR sources driven by higher-frequency microwave sources exist specifically to push charge states and beam energies higher, entirely separate from any change to the cyclotron's own magnet.

**This is worth highlighting in your presentation**: your group can propose a charge-state upgrade (a new or improved ECR source) as an alternative, or a complement, to a radius upgrade. As Part 3 shows, a fairly modest increase in achievable charge state can rival a geometrically much larger, more expensive radius expansion.

## Part 5: What Does That Extra Energy Actually Do?

A projectile needs enough kinetic energy to overcome the **Coulomb barrier**, the electrostatic repulsion between the positively charged projectile and target nucleus, before the strong nuclear force can take over and fuse them together. We compare the bismuth beam's energy, before and after each kind of upgrade, to the Coulomb barrier for a symmetric bismuth-on-bismuth collision.

```python
def coulomb_barrier(Z1, A1, Z2, A2):
    """Approximate Coulomb barrier in MeV, using R = 1.2*(A1^(1/3) + A2^(1/3)) fm."""
    R = 1.2 * (A1**(1/3) + A2**(1/3))
    return 1.44 * Z1 * Z2 / R

Vc_bibi = coulomb_barrier(Z_bi, A_bi, Z_bi, A_bi)

excess_current = E_total_current - Vc_bibi
excess_new      = E_total_new - Vc_bibi

print(f"Coulomb barrier, bismuth-209 on bismuth-209: {Vc_bibi:.1f} MeV")
print()
print(f"Current beam total energy:  {E_total_current:.1f} MeV  (excess above barrier: {excess_current:.1f} MeV)")
print(f"Upgraded beam total energy: {E_total_new:.1f} MeV  (excess above barrier: {excess_new:.1f} MeV)")
print()
print("Both beams already clear this particular barrier. So the upgrade is not a story of")
print("'newly crossing the barrier' for this reaction. It is a story of how far above the")
print("barrier you land, which is what actually matters for the next step.")
```

That excess energy above the barrier becomes excitation energy in the compound nucleus formed by fusion, and it is exactly what controls how many neutrons boil off afterward. More excess energy means more neutrons evaporate, which means a different final isotope is produced. This is the real, tangible thing the upgrade changes: not whether fusion happens, but which isotope comes out the other side.

## Part 6: Try Your Own Beam and Target

The same method works for any beam-target pair, and for either lever. Change the values below to explore a different combination, and try raising `assumed_Q_beam` instead of changing the radius to see the charge-state lever in action for your own choice.

```python
ions = {
    'Ca-48':    (20, 48),
    'Ti-50':    (22, 50),
    'Ni-64':    (28, 64),
    'Cm-248':   (96, 248),
    'Cf-249':   (98, 249),
}

# ── Change these two lines to explore a different combination ──────────────
beam_name, target_name = 'Ca-48', 'Cf-249'
assumed_Q_beam = 12   # an assumed, illustrative achievable charge state for the beam ion

Z_beam, A_beam = ions[beam_name]
Z_targ, A_targ = ions[target_name]

Vc = coulomb_barrier(Z_beam, A_beam, Z_targ, A_targ)

E_per_A_88_custom  = K_88  * (assumed_Q_beam / A_beam)**2
E_per_A_100_custom = K_100 * (assumed_Q_beam / A_beam)**2
E_total_88_custom  = E_per_A_88_custom  * A_beam
E_total_100_custom = E_per_A_100_custom * A_beam

print(f"Beam: {beam_name} (Z={Z_beam}, A={A_beam}), assumed Q = {assumed_Q_beam}+")
print(f"Target: {target_name} (Z={Z_targ}, A={A_targ})")
print(f"Coulomb barrier: {Vc:.1f} MeV")
print()
print(f"Current (K={K_88}):  {E_total_88_custom:.1f} MeV total  (excess: {E_total_88_custom - Vc:.1f} MeV)")
print(f"Upgraded (K={K_100:.0f}): {E_total_100_custom:.1f} MeV total  (excess: {E_total_100_custom - Vc:.1f} MeV)")
print()
if E_total_88_custom < Vc <= E_total_100_custom:
    print("This reaction newly crosses the barrier because of the upgrade!")
elif E_total_88_custom >= Vc:
    print("This reaction already clears the barrier today; the upgrade increases your excess energy.")
else:
    print("This reaction still does not reach the barrier even after the upgrade.")
```

## Part 7: Would the Detectors Also Need to Change?

A bigger magnet does not just change what reactions are possible; it changes what needs to catch and identify the reaction products afterward.

- **Higher-energy reaction products** may need thicker or denser detector material to fully stop and measure their energy
- **Particle identification** (telling reaction products apart by species) may need to be recalibrated, since higher-energy particles behave differently in silicon detectors and scintillators than the lower-energy products your current setup was designed for
- **Shielding** may need to be reconsidered, since higher-energy reactions can produce more penetrating secondary radiation (neutrons, gammas), a safety consideration as much as a physics one
- **Beam time and rates** may also shift: a higher-energy beam interacting with a target can change reaction rates, requiring faster or more radiation-hard electronics
