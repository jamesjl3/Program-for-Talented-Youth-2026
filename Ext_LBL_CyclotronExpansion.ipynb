## Part 1: From Radius to Momentum

We treat the cyclotron as a single effective bending radius (a simplified model; a real cyclotron's spiral path is more complex, but the same scaling holds). We compare the current 88-inch magnet to the proposed 100-inch magnet, holding the magnetic field fixed.

```python
# ── Geometry: convert named radius (inches) to meters ──────────────────────────
# The "88-inch" and "100-inch" names refer to the magnet pole diameter; we take
# half of that as the effective bending radius available to the beam.
in_to_m = 0.0254

r_88_m  = (88 / 2) * in_to_m
r_100_m = (100 / 2) * in_to_m

B_field = 1.5   # tesla, illustrative typical operating field for this class of cyclotron

# ── Proton beam (q = 1e, mass = 0.938 GeV/c^2) ─────────────────────────────────
q_proton = 1
m_proton = 0.938   # GeV/c^2

def momentum_GeV(q, B_T, r_m):
    """p [GeV/c] = 0.2998 * q * B[T] * r[m], for charge q in units of e."""
    return 0.2998 * q * B_T * r_m

def kinetic_energy_GeV(p_GeV, m_GeV):
    """Relativistic kinetic energy from momentum and rest mass."""
    return np.sqrt(p_GeV**2 + m_GeV**2) - m_GeV

p_88  = momentum_GeV(q_proton, B_field, r_88_m)
p_100 = momentum_GeV(q_proton, B_field, r_100_m)

KE_88_MeV  = kinetic_energy_GeV(p_88, m_proton) * 1000
KE_100_MeV = kinetic_energy_GeV(p_100, m_proton) * 1000

print(f"Effective bending radius, 88-inch magnet:  {r_88_m:.3f} m")
print(f"Effective bending radius, 100-inch magnet: {r_100_m:.3f} m")
print(f"Radius increase: {(r_100_m/r_88_m - 1)*100:.1f}%")
print()
print(f"Proton kinetic energy, 88-inch magnet:  {KE_88_MeV:.1f} MeV")
print(f"Proton kinetic energy, 100-inch magnet: {KE_100_MeV:.1f} MeV")
print(f"Energy increase: {(KE_100_MeV/KE_88_MeV - 1)*100:.1f}%")
```

Notice that the kinetic energy increases by a larger percentage than the radius did. At these energies, kinetic energy depends on momentum roughly as $p^2$, so a modest gain in radius (and therefore momentum) turns into a bigger relative gain in energy.

## Part 2: What Does That Extra Energy Buy You?

In nuclear reaction studies, a projectile needs enough kinetic energy to overcome the **Coulomb barrier**, the electrostatic repulsion between the positively charged projectile and target nucleus, before the strong nuclear force can take over and fuse them together.

We compare your two available beam energies against three schematic reaction barriers (illustrative values, not specific published reactions) to see which reactions newly become reachable with the bigger magnet.

```python
reactions = [
    ("Reaction A (lighter target)",  100),
    ("Reaction B (medium target)",   140),
    ("Reaction C (heavier target)",  175),
]

print(f"{'Reaction':32s} {'Barrier (MeV)':>14s} {'Accessible at 88in?':>22s} {'Accessible at 100in?':>22s}")
print("-" * 92)
for name, barrier in reactions:
    ok_88  = "Yes" if KE_88_MeV  > barrier else "No"
    ok_100 = "Yes" if KE_100_MeV > barrier else "No"
    print(f"{name:32s} {barrier:14.0f} {ok_88:>22s} {ok_100:>22s}")
```

```python
fig, ax = plt.subplots(figsize=(9, 5))

names, barriers = zip(*reactions)
y_pos = np.arange(len(names))

ax.barh(y_pos, barriers, color='gray', alpha=0.5, label='Coulomb barrier')
ax.axvline(x=KE_88_MeV, color='steelblue', linewidth=2.5, label=f'88-inch beam energy ({KE_88_MeV:.0f} MeV)')
ax.axvline(x=KE_100_MeV, color='firebrick', linewidth=2.5, label=f'100-inch beam energy ({KE_100_MeV:.0f} MeV)')
ax.set_yticks(y_pos)
ax.set_yticklabels(names)
ax.set_xlabel('Energy (MeV)', fontsize=12)
ax.set_title('Which reactions does the bigger magnet newly unlock?', fontsize=12)
ax.legend(fontsize=10, loc='lower right')
ax.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()
```

## Part 3: Would the Detectors Also Need to Change?

A bigger magnet does not just change what reactions are possible; it changes what needs to catch and identify the reaction products afterward.

- **Higher-energy reaction products** may need thicker or denser detector material to fully stop and measure their energy
- **Particle identification** (telling reaction products apart by species) may need to be recalibrated, since higher-energy particles behave differently in silicon detectors and scintillators than the lower-energy products your current setup was designed for
- **Shielding** may need to be reconsidered, since higher-energy reactions can produce more penetrating secondary radiation (neutrons, gammas), a safety consideration as much as a physics one
- **Beam time and rates** may also shift: a higher-energy beam interacting with a target can change reaction rates, requiring faster or more radiation-hard electronics
