In Act 2, we smashed two protons together at CERN and rediscovered the Z boson. Now we go to a completely different machine and a completely different regime of physics.

At the **Relativistic Heavy Ion Collider (RHIC)** at Brookhaven National Laboratory on Long Island, New York, physicists accelerate gold nuclei (¹ₗₗAu, 79 protons, 118 neutrons each) to 99.995% of the speed of light and collide them head-on.

The energy density reached in these collisions exceeds **10 GeV/fm³**, about 50 times denser than normal nuclear matter. Under these conditions, protons and neutrons dissolve. Quarks and gluons are no longer confined inside individual hadrons; instead, they roam freely in a state called the **Quark-Gluon Plasma (QGP)**.

This is the state the entire universe was in for the first **10 microseconds** after the Big Bang.

---

### How do we know the QGP forms?

The QGP itself lives for only ~5 fm/c (~10⁻²³ seconds) before cooling and re-forming into hadrons. We can't see it directly. Instead, we look at **signatures**, things that change between p+p and Au+Au collisions.

One of the most powerful signatures is **jet quenching**. High-energy quarks and gluons produced in the collision must travel through the QGP medium. As they do, they lose energy, like a bullet passing through water instead of air. This energy loss **suppresses** the production of high-pT particles.

We quantify this with the **nuclear modification factor R_AA**:

$$R_{AA}(pT) = \frac{\text{yield in Au+Au}}{\langle T_{AA} \rangle \times \text{yield in p+p}}$$

Where $\langle T_{AA} \rangle$ (the nuclear overlap function) accounts for the fact that an Au+Au collision contains many independent nucleon-nucleon collisions.

- **R_AA = 1** → Au+Au looks just like scaled-up p+p → no medium effect
- **R_AA < 1** → suppression → particles are being absorbed or slowed by the QGP
- **R_AA > 1** → enhancement → something is producing *more* particles than expected

In this act you will compute and visualize R_AA from **real published PHENIX data** at RHIC.

## Part 1: Loading the Data

The data below comes from the **PHENIX experiment at RHIC**, the same experiment that sPHENIX is the successor to. These are published results from:

We have three datasets:

2. **Au+Au central (0–10%)** - head-on collisions: maximum overlap, hottest fireball
3. **Au+Au peripheral (60–92%)** - peripheral collisions: small overlap, less medium

```python
pp_pT    = np.array([0.375, 0.625, 0.875, 1.25, 1.75, 2.25, 2.75, 3.25, 3.75, 4.25, 4.75])
pp_yield = np.array([2.27e1, 5.47, 1.56, 2.85e-1, 3.38e-2, 5.64e-3, 1.08e-3, 2.27e-4, 5.25e-5, 1.30e-5, 3.52e-6])
pp_err   = np.array([0.22e1, 0.30, 0.09, 0.21e-1, 0.25e-2, 0.42e-3, 0.09e-3, 0.21e-4, 0.58e-5, 0.17e-5, 0.55e-6])

auau_central_pT    = np.array([0.375, 0.625, 0.875, 1.25, 1.75, 2.25, 2.75, 3.25, 3.75, 4.25, 4.75])
auau_central_yield = np.array([7.42e2, 1.28e2, 3.06e1, 4.06, 3.77e-1, 4.87e-2, 7.65e-3, 1.31e-3, 2.49e-4, 5.24e-5, 1.20e-5])
auau_central_err   = np.array([0.12e2, 0.04e2, 0.08e1, 0.10, 0.09e-1, 0.11e-2, 0.19e-3, 0.04e-3, 0.08e-4, 0.17e-5, 0.05e-5])
T_AA_central = 22.7

auau_periph_pT    = np.array([0.375, 0.625, 0.875, 1.25, 1.75, 2.25, 2.75, 3.25, 3.75, 4.25, 4.75])
auau_periph_yield = np.array([7.440e+00, 1.736e+00, 4.736e-01, 8.358e-02, 9.562e-03, 1.557e-03, 2.906e-04, 5.952e-05, 1.358e-05, 3.319e-06, 8.865e-07])
auau_periph_err   = np.array([0.15e+00, 0.04e+00, 0.10e-01, 0.16e-02, 0.18e-03, 0.30e-04, 0.06e-04, 0.15e-05, 0.03e-05, 0.08e-06, 0.02e-06])
T_AA_periph = 0.345

print('PHENIX data loaded!')
print(f'  p+p:          {len(pp_pT)} pT points, covering {pp_pT[0]:.2f}–{pp_pT[-1]:.2f} GeV/c')
print(f'  Au+Au central:{len(auau_central_pT)} pT points')
print(f'  Au+Au periph: {len(auau_periph_pT)} pT points')
```

## Part 2: Comparing pT Spectra: p+p vs Au+Au

The first thing to look at: how does the **shape** of the pT spectrum change between p+p and Au+Au?

To make a fair comparison, we need to scale the Au+Au yield by T_AA. This tells us what we'd *expect* in Au+Au if it were simply many independent p+p collisions with no medium effects.

```python
pp_scaled_central = pp_yield * T_AA_central
pp_scaled_periph  = pp_yield * T_AA_periph

fig, ax = plt.subplots(figsize=(8, 6))
ax.errorbar(auau_central_pT, auau_central_yield, yerr=auau_central_err,
            fmt='o', color='firebrick', markersize=7, capsize=4,
            label='Au+Au central (0–10%): head-on', zorder=3)
ax.errorbar(auau_periph_pT, auau_periph_yield, yerr=auau_periph_err,
            fmt='s', color='steelblue', markersize=7, capsize=4,
            label='Au+Au peripheral (60–92%): peripheral', zorder=3)
ax.plot(pp_pT, pp_scaled_central, 'r--', linewidth=1.5, alpha=0.7,
        label='T_AA × p+p (central expectation, no QGP)')
ax.plot(pp_pT, pp_scaled_periph,  'b--', linewidth=1.5, alpha=0.7,
        label='T_AA × p+p (peripheral expectation, no QGP)')
ax.set_yscale('log')
ax.set_xlabel('Transverse Momentum pT (GeV/c)', fontsize=13)
ax.set_ylabel('Invariant yield (GeV/c)⁻²', fontsize=13)
ax.set_title('Charged hadron pT spectra: p+p vs Au+Au at √s_NN = 200 GeV\n'
             'PHENIX experiment at RHIC (Adler et al., PRL 91, 072301, 2003)', fontsize=11)
ax.legend(fontsize=10, loc='upper right')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print('The dashed lines show what Au+Au should look like if there were no QGP.')
print('Look carefully at high pT (right side). What do you notice about the central data?')
```

**What do you see?** Both Au+Au datasets fall **below** their dashed expectations at high pT. Suppression is happening in both central and peripheral collisions. But look more carefully: the **central** data falls further below its line than the **peripheral** data does. The suppression is stronger when more nuclei overlap.

This is the signature of **jet quenching** where high-energy partons lose energy as they travel through the hot, dense medium. The more QGP is created (more central = more overlap), the more energy is lost.

## Part 3: Computing R_AA

The ratio R_AA makes the comparison precise. Let's compute it step by step.

```python
RAA_central = auau_central_yield / (T_AA_central * pp_yield)
RAA_periph  = auau_periph_yield  / (T_AA_periph  * pp_yield)

RAA_central_err = RAA_central * np.sqrt((auau_central_err/auau_central_yield)**2 + (pp_err/pp_yield)**2)
RAA_periph_err  = RAA_periph  * np.sqrt((auau_periph_err /auau_periph_yield )**2 + (pp_err/pp_yield)**2)

print('R_AA values: Central Au+Au (0-10%):')
for pT, raa in zip(auau_central_pT, RAA_central):
    bar = '█' * int(raa * 20)
    print(f'  pT = {pT:.2f} GeV/c:  R_AA = {raa:.3f}  {bar}')
print()
print('R_AA = 1.0 would mean no suppression.')
print('Values below 1 indicate suppression (jet quenching).')
```

## Part 4: Plotting R_AA: The Smoking Gun

```python
fig, ax = plt.subplots(figsize=(9, 6))
ax.axhline(y=1.0, color='black', linestyle='-', linewidth=1.5,
           label='R_AA = 1 (no medium effect)')
ax.axhspan(0.9, 1.1, alpha=0.08, color='black')
ax.errorbar(auau_central_pT, RAA_central, yerr=RAA_central_err,
            fmt='o', color='firebrick', markersize=9, capsize=5, linewidth=1.5,
            label='Au+Au central (0–10%): head-on collision', zorder=4)
ax.errorbar(auau_periph_pT, RAA_periph, yerr=RAA_periph_err,
            fmt='s', color='steelblue', markersize=9, capsize=5, linewidth=1.5,
            label='Au+Au peripheral (60–92%): peripheral collision', zorder=4)
ax.annotate('Suppression!\n(jet quenching)',
            xy=(3.5, RAA_central[8]), xytext=(3.8, 0.55),
            fontsize=11, color='firebrick',
            arrowprops=dict(arrowstyle='->', color='firebrick', lw=1.5))
ax.set_xlabel('Transverse Momentum pT (GeV/c)', fontsize=13)
ax.set_ylabel('Nuclear Modification Factor  R_AA', fontsize=13)
ax.set_title('Jet quenching in Au+Au collisions at RHIC (√s_NN = 200 GeV)\n'
             'PHENIX experiment - Adler et al., PRL 91, 072301 (2003)', fontsize=11)
ax.set_ylim(0, 1.8)
ax.set_xlim(0, 5.2)
ax.legend(fontsize=11, loc='upper left')
ax.grid(True, alpha=0.3)
textstr = 'Central Au+Au: R_AA ≈ 0.2 at high pT\n→ 80% suppression of high-energy particles\n→ Signature of the Quark-Gluon Plasma'
ax.text(0.02, 0.25, textstr, transform=ax.transAxes, fontsize=9,
        verticalalignment='bottom', bbox=dict(boxstyle='round', facecolor='mistyrose', alpha=0.8))
plt.tight_layout()
plt.show()
```

```python
high_pT_mask = auau_central_pT >= 2.5
avg_RAA_high_pT = np.mean(RAA_central[high_pT_mask])
print(f'Average R_AA for pT ≥ 2.5 GeV/c in central Au+Au:')
print(f'  R_AA = {avg_RAA_high_pT:.2f}')
print(f'  This means {100*(1-avg_RAA_high_pT):.0f}% of high-pT particles are suppressed.')
print()
print('For comparison, peripheral Au+Au at the same pT range:')
avg_RAA_periph = np.mean(RAA_periph[high_pT_mask])
print(f'  R_AA = {avg_RAA_periph:.2f}')
print(f'  Only {100*(1-avg_RAA_periph):.0f}% suppression; much less QGP formed.')
```

Think of it like throwing two basketballs at each other:
- **Central (0%)** = direct hit → huge explosion, maximum overlap, most QGP

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
centrality_labels = ['Central (0–10%)\nDirect hit', 'Mid-central (20–40%)\nModerate overlap', 'Peripheral (60–92%)\nGrazing blow']
colors = ['firebrick', 'darkorange', 'steelblue']
b_values = [0.15, 0.55, 0.88]
R = 1.0
import numpy as np
theta = np.linspace(0, 2*np.pi, 300)
for ax, label, color, b in zip(axes, centrality_labels, colors, b_values):
    ax.fill(R*np.cos(theta) - b*R, R*np.sin(theta), alpha=0.4, color=color)
    ax.plot(R*np.cos(theta) - b*R, R*np.sin(theta), color=color, linewidth=2)
    ax.fill(R*np.cos(theta) + b*R, R*np.sin(theta), alpha=0.4, color='gray')
    ax.plot(R*np.cos(theta) + b*R, R*np.sin(theta), color='gray', linewidth=2)
    ax.annotate('', xy=(b*R, 0), xytext=(-b*R, 0),
                arrowprops=dict(arrowstyle='<->', color='black', lw=1.5))
    ax.text(0, -0.25, f'b = {b:.2f} x 2R', ha='center', fontsize=9)
    ax.set_title(label, fontsize=10, color=color)
    ax.set_xlim(-2.5, 2.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
plt.suptitle('Collision centrality: how head-on is the collision?\n'
             '(View along the beam axis: the two nuclei come from above and below)',
             fontsize=11, y=1.05)
plt.tight_layout()
plt.show()
```

## Extension: The Particles That Got Away: Neutrinos and Missing Transverse Energy

So far, you have measured R_AA for charged hadrons, particles that *do* interact with the detector. But not everything that comes out of a gold-gold collision gets caught.

**Neutrinos** pass through the entire detector without leaving a trace. Every collision produces them, and PHENIX (and sPHENIX) cannot see them directly.

How do physicists know they were there?

### Conservation of momentum: The missing transverse energy trick

Before the collision, the two gold nuclei travel head-on. The total momentum **perpendicular to the beam** (the *transverse* direction) is exactly **zero**.

After the collision, if you add up the transverse momentum of every particle you *can* detect, momentum conservation says the sum should still be zero.

If it is **not** zero (if there is a momentum imbalance), then something invisible carried it away.

That imbalance is called **Missing Transverse Energy (MET)**, or sometimes $\not\!E_T$.

$$\vec{p}_{\,\text{missing}} = -\sum_{\text{detected}} \vec{p}_{T,\,i}$$

In proton-proton collisions at the LHC, MET is one of the most powerful signals for new physics, including W bosons decaying to neutrinos, and searches for dark matter.

In heavy-ion collisions at RHIC, neutrinos escape even the densest QGP fireballs. Their R_AA is expected to be **exactly 1**, meaning they do not interact with the QGP at all. That makes them a perfect control: any suppression you see in hadrons cannot be blamed on geometric effects, because neutrinos show none.



```python
# (numpy and matplotlib already loaded above)


# --- Toy model: a single Au+Au collision ---
# We simulate detected particles (pions, kaons, protons)
# and one invisible neutrino.

np.random.seed(42)
n_detected = 200  # charged hadrons we can see

# Random transverse momenta for detected particles (GeV/c)
# Drawn from a thermal-like distribution
pt_mag  = np.random.exponential(scale=0.4, size=n_detected)  # magnitude
phi     = np.random.uniform(0, 2*np.pi, size=n_detected)     # azimuthal angle
px_det  = pt_mag * np.cos(phi)
py_det  = pt_mag * np.sin(phi)

# The neutrino carries away some momentum
nu_px, nu_py = 1.8, -1.2  # GeV/c which represents the "invisible" momentum

# Total detected imbalance
MET_x = -(px_det.sum())   # what the detector infers as missing
MET_y = -(py_det.sum())
MET   = np.sqrt(MET_x**2 + MET_y**2)

# True MET (includes neutrino)
MET_x_true = -(px_det.sum() + nu_px)
MET_y_true = -(py_det.sum() + nu_py)
MET_true   = np.sqrt(MET_x_true**2 + MET_y_true**2)

print(f"Sum of detected px:  {px_det.sum():.3f} GeV/c")
print(f"Sum of detected py:  {py_det.sum():.3f} GeV/c")
print(f"")
print(f"MET inferred from detector: {MET:.3f} GeV/c  (should be ~0 if nothing invisible)")
print(f"True MET (neutrino included): {MET_true:.3f} GeV/c")
print(f"")
print(f"The neutrino carried:  px={nu_px:.1f}, py={nu_py:.1f} GeV/c")
print(f"The residual MET points back at it!")
```

```python
# --- Visualize the transverse plane ---
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for ax, title, show_nu in zip(axes,
    ['What the detector sees', 'True picture (neutrino included)'],
    [False, True]):

    # Draw detected particle momenta as small arrows
    for i in range(n_detected):
        ax.annotate('', xy=(px_det[i], py_det[i]), xytext=(0, 0),
                    arrowprops=dict(arrowstyle='->', color='steelblue', lw=0.4, alpha=0.5))

    # Draw the vector sum of detected particles
    ax.annotate('', xy=(px_det.sum(), py_det.sum()), xytext=(0, 0),
                arrowprops=dict(arrowstyle='->', color='crimson', lw=2.5))
    ax.text(px_det.sum()*1.1, py_det.sum()*1.1,
            r'$\Sigma\vec{p}_T$ (detected)', color='crimson', fontsize=10)

    if show_nu:
        ax.annotate('', xy=(nu_px, nu_py), xytext=(0, 0),
                    arrowprops=dict(arrowstyle='->', color='gold', lw=2.5,
                                   linestyle='dashed'))
        ax.text(nu_px*1.15, nu_py*1.15, r'$\nu$ (invisible!)',
                color='goldenrod', fontsize=11, fontweight='bold')

        ax.annotate('', xy=(px_det.sum()+nu_px, py_det.sum()+nu_py), xytext=(0, 0),
                    arrowprops=dict(arrowstyle='->', color='gray', lw=1.5,
                                   linestyle='dotted'))
        ax.text(px_det.sum()+nu_px+0.1, py_det.sum()+nu_py,
                'True sum\n(~0)', color='gray', fontsize=9)

    ax.axhline(0, color='black', lw=0.5, alpha=0.4)
    ax.axvline(0, color='black', lw=0.5, alpha=0.4)
    ax.set_xlim(-4, 4); ax.set_ylim(-4, 4)
    ax.set_xlabel(r'$p_x$ (GeV/c)', fontsize=12)
    ax.set_ylabel(r'$p_y$ (GeV/c)', fontsize=12)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.2)

plt.suptitle('Transverse Momentum Balance in a Heavy-Ion Collision',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
print("Left panel: the detector sees an imbalance. Does this mean momentum is not conserved?")
print("Right panel: the neutrino explains everything. Momentum IS conserved.")
```

### Why can't PHENIX (or sPHENIX) detect neutrinos?

Neutrinos interact only via the **weak nuclear force**, which has an incredibly tiny cross-section. A neutrino can pass through a light-year of lead with only a 50% chance of interacting.

PHENIX and sPHENIX are built around electromagnetic and hadronic calorimeters, where they catch particles that interact electromagnetically or via the strong force. Neutrinos do neither.

---

### Enter IceCube

IceCube takes the opposite approach. Instead of a compact detector surrounded by calorimeters, it uses **one cubic kilometer of Antarctic ice** as the detector medium.

When a very high-energy neutrino (TeV-scale) finally does interact in the ice, it produces a secondary charged particle (a **muon** or a shower of particles) that travels faster than light moves *through ice* (though not faster than light in vacuum). This produces **Cherenkov radiation**: a cone of blue light, the optical equivalent of a sonic boom.

5,160 photomultiplier tubes buried in the ice detect that light cone and reconstruct the neutrino's direction and energy.

The key insight for the IceCube project: **IceCube detects neutrinos that accelerators cannot**. If you placed an accelerator above IceCube, the neutrino beam would pass straight through the Earth and be detected on the other side or interact in the ice below. This is a real technique called a **neutrino beam experiment**.

---

```python
# --- Neutrino production rate: why accelerators matter ---
# Neutrinos from astrophysical sources (AGN, supernovae) are diffuse and rare.
# A controlled accelerator beam has a KNOWN direction, energy, and flux.
# That turns IceCube from a telescope into a precision instrument.

# Toy comparison of neutrino fluxes
sources = [
    ('Atmospheric\n(cosmic ray)\nbackground',   1e4,  'steelblue'),
    ('Astrophysical\n(AGN, diffuse)',            3e2,  'darkorange'),
    ('Accelerator beam\n(hypothetical,\npointed at IceCube)', 1e6, 'crimson'),
]

fig, ax = plt.subplots(figsize=(8, 5))
labels  = [s[0] for s in sources]
fluxes  = [s[1] for s in sources]
colors  = [s[2] for s in sources]

bars = ax.bar(labels, fluxes, color=colors, edgecolor='black', linewidth=0.8)
ax.set_yscale('log')
ax.set_ylabel(r'Relative $\nu$ flux (arbitrary units, log scale)', fontsize=11)
ax.set_title('Neutrino Flux at IceCube: Natural vs. Accelerator Sources',
             fontsize=12, fontweight='bold')
ax.tick_params(axis='x', labelsize=10)
ax.grid(axis='y', alpha=0.3)

for bar, flux in zip(bars, fluxes):
    ax.text(bar.get_x() + bar.get_width()/2, flux*1.5,
            f'{flux:.0e}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

print("An accelerator beam pointed at IceCube would produce a neutrino flux")
print("orders of magnitude larger than natural sources and with known energy and direction.")
print("That is the key advantage of their group's proposed design.")
```

### Exploration 3: The connection to sPHENIX

The data you just analyzed came from **PHENIX**, which ran at RHIC from 2000 to 2016. PHENIX measured R_AA like this for the first time, proving the QGP exists.

**sPHENIX** is its successor; a brand-new detector that took its first Au+Au data in 2023. sPHENIX was specifically designed to measure **jet substructure** inside the QGP: not just *how many* high-pT particles are suppressed, but *how* the internal structure of jets changes as they pass through the QGP fireball.

That's what your instructor's thesis research is about.

```python
z = np.linspace(0.05, 0.95, 200)
ff_pp   = (1 - z)**4 / z
suppression = np.where(z > 0.3, 1 - 0.6 * ((z - 0.3) / 0.7)**0.5, 1.0)
enhancement = np.where(z < 0.2, 1 + 0.8 * (0.2 - z) / 0.2, 1.0)
ff_auau = ff_pp * suppression * enhancement

plt.figure(figsize=(9, 5))
plt.semilogy(z, ff_pp,   color='black',     linewidth=2.5, linestyle='-',  label='p+p (vacuum, no QGP)')
plt.semilogy(z, ff_auau, color='firebrick', linewidth=2.5, linestyle='--', label='Au+Au (with QGP)')
plt.fill_between(z, ff_auau, ff_pp,
                 where=(ff_pp > ff_auau), alpha=0.25, color='firebrick', label='Suppressed (high-z)')
plt.fill_between(z, ff_auau, ff_pp,
                 where=(ff_auau > ff_pp), alpha=0.25, color='steelblue', label='Enhanced (low-z)')
plt.axvline(x=0.3, color='gray', linestyle=':', linewidth=1.2, alpha=0.7)
plt.text(0.31, 0.15, 'suppression\nstarts here', fontsize=9, color='gray')
plt.xlabel('Momentum fraction  z = pT_fragment / pT_jet', fontsize=12)
plt.ylabel('Fragmentation function D(z)  [log scale]', fontsize=12)
plt.title('Toy model: How jet fragmentation changes in the QGP\n'
          '(This is what sPHENIX is measuring right now!)', fontsize=11)
plt.legend(fontsize=10, loc='upper right')
plt.grid(True, alpha=0.3, which='both')
plt.tight_layout()
plt.show()

print('High-z fragments (hard, carry most of the jet momentum) are suppressed in the QGP.')
print('Low-z fragments (soft) are enhanced (the lost energy goes somewhere!)')
print('sPHENIX measures this precisely using jet substructure observables like Rg and zg.')
```

Over the last three acts, you've traveled the full arc of modern nuclear physics:

| Act | System | What we learned |
|---|---|---|
| **1** | e+p (HERA/PYTHIA) | Protons are made of quarks: up and down valence quarks plus a sea |
| **2** | p+p (LHC/CMS) | When two protons collide, we can rediscover the Z boson and other particles from invariant mass peaks |

### The open questions your instructor is working on right now:

The PHENIX measurement showed *that* jets are quenched. But **how**? What is the microscopic mechanism?

- Does a jet lose energy by radiating gluons? By colliding with thermal quarks in the medium?
- Does the internal **structure** of a jet change, or does it become wider? More or less collimated?
- Do coherence effects (where parts of a jet look like a single particle to the medium) play a role?

These are exactly the questions that **sPHENIX** was built to answer; using observables like the groomed jet radius (Rg) and the momentum splitting fraction (zg) that probe the jet's internal structure before and after it passes through the QGP.
