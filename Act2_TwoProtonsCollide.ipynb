Now we go bigger.At CERN's **Large Hadron Collider (LHC)** near Geneva, Switzerland, two beams of protons are accelerated in opposite directions around a 27 km circular tunnel and smashed together at nearly the speed of light. The collision energy is so extreme that entirely new, massive particles can be created from pure energy: particles that haven't existed freely in the universe since fractions of a second after the Big Bang.In this act, you will load **real data** from the **CMS detector** at the LHC: 100,000 actual proton-proton collisions recorded in 2010: and use it to **rediscover known particles** hiding in the data.-### The key idea: Invariant MassMany heavy particles are too short-lived to reach a detector: they decay almost instantly into lighter particles. The **Z boson**, for example, lives for only about 3 × 10⁻²⁵ seconds before decaying into two muons (μ⁺μ⁻).We can't see the Z directly, but we can reconstruct its mass from the two muons it leaves behind using **special relativity**:$$M_{inv} = \sqrt{(E_1 + E_2)^2 - (\vec{p}_1 + \vec{p}_2)^2}$$If many events produce a pair of muons with the same reconstructed mass, that mass corresponds to a **real particle**. This is how particle physicists discover new particles: by finding peaks in the invariant mass spectrum.**The dataset has already computed M for us.** Let's load it and see what's there.-First, load our libraries:

```python
import numpy as npimport pandas as pdimport matplotlib.pyplot as pltprint('Libraries loaded!')
```

-## Part 1: Loading Real CMS DataThe data comes from the **CERN Open Data Portal**: a public repository where CERN releases real collision data for education and research. We load it directly from the web with one line of pandas.

```python
# Load 100,000 real dimuon events from the CMS experiment at CERN# These are proton-proton collisions at sqrt(s) = 7 TeV, recorded in 2010# Source: CERN Open Data Portal, record 700 (CC0 public domain)print('Loading CMS data from CERN Open Data Portal...')url = 'http://opendata.cern.ch/record/700/files/MuRun2010B.csv'cms = pd.read_csv(url)print(f'Loaded {len(cms):,} events!')print(f'Columns: {list(cms.columns)}')
```

```python
# Look at the first few rowscms.head(10)
```

### What are we looking at?Each row is one collision event where **exactly two muons** were detected. The columns come in pairs: one for each muon:| Column | Meaning ||-|-|| `E1`, `E2` | Energy of muon 1 and muon 2 (GeV) || `pt1`, `pt2` | Transverse momentum (GeV/c) || `eta1`, `eta2` | Pseudorapidity (angle from beam) || `phi1`, `phi2` | Azimuthal angle (radians) || `Q1`, `Q2` | Electric charge (+1 or −1) || **`M`** | **Invariant mass of the muon pair (GeV/c²): this is the key column** |These muons were measured by the CMS detector: a 14,000-tonne instrument the size of a five-story building buried 100 meters underground.

```python
# Basic statistics on the invariant massprint('Invariant mass summary:')print(f' Min: {cms["M"].min():.2f} GeV/c²')print(f' Max: {cms["M"].max():.2f} GeV/c²')print(f' Mean: {cms["M"].mean():.2f} GeV/c²')print(f'\nFor reference: proton mass = 0.938 GeV/c²')print(f' Z boson mass = 91.2 GeV/c²')
```

-## Part 2: The Full Invariant Mass SpectrumNow the central question: if we plot a histogram of the invariant mass `M` for all 100,000 events, what do we see?If a real particle decays into two muons, events with that invariant mass will pile up: creating a **peak** at that particle's mass. The smoother background comes from random muon pairs that don't come from a single parent particle.Let's look at the full range first:

```python
plt.figure(figsize=(10, 5))plt.hist(cms['M'], bins=500, color='royalblue', alpha=0.8)plt.xlabel('Dimuon Invariant Mass M (GeV/c²)', fontsize=13)plt.ylabel('Number of events', fontsize=13)plt.title('Full dimuon invariant mass spectrum: 100,000 real CMS collisions (2010)', fontsize=12)plt.yscale('log') # log scale to see features across many orders of magnitudeplt.tight_layout()plt.show()print('Tip: the y-axis is logarithmic: each gridline is 10x larger than the one below it.')
```

**Can you see any peaks?** There are several hiding in there. Let's zoom in to find them one by one.

-## Part 3: Particle Hunting: Zoom In### Peak 1: The J/ψ meson (~3.1 GeV)The **J/ψ** ("J-psi") is a bound state of a charm quark and an anti-charm quark. Its discovery in 1974: simultaneously at BNL and SLAC: was so dramatic it was called the **"November Revolution"** in particle physics. Both teams shared the 1976 Nobel Prize.

```python
# Zoom into the J/psi region: 2 to 5 GeVjpsi_region = cms[(cms['M'] > 2) & (cms['M'] < 5)]plt.figure(figsize=(8, 4))plt.hist(jpsi_region['M'], bins=150, color='darkorange', alpha=0.85, edgecolor='black', linewidth=0.3)plt.axvline(x=3.097, color='red', linestyle='--', linewidth=1.5, label='J/ψ PDG mass: 3.097 GeV/c²')plt.xlabel('Dimuon Invariant Mass M (GeV/c²)', fontsize=13)plt.ylabel('Number of events', fontsize=13)plt.title('J/ψ region (2-5 GeV): charm-anticharm bound state', fontsize=12)plt.legend(fontsize=11)plt.tight_layout()plt.show()
```

The same-charge background is flat: no peak. The Z boson peak only appears in opposite-charge pairs, exactly as expected. **This is how physicists separate signal from background.**
